```python
# The number of matches played in each IPL season
plt.figure(figsize = (10, 6))
sns.countplot(x = 'season', data = matches)
plt.xlabel('season')
plt.show()
```

```python
# The number of matches won by each team
plt.figure(figsize = (10, 6))
sns.countplot(x = 'winner', data = matches, palette = 'crest')
plt.title('The number of matches won by each team')
plt.xlabel('Teams')
plt.ylabel('Number of wins')
plt.show()
```

```python
matches['win_by'] = np.where(matches['win_by_runs'] > 0, 'Bat first', 'Bowl first')
```

```python
plt.figure(figsize = (10, 6))
sns.countplot(x = 'season', hue = 'win_by', data = matches, palette = 'rocket_r')
plt.title('Match wins by batting and bowling')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
# The number of matches played in the stadium
plt.figure(figsize = (10, 6))
plt.xticks(rotation = 90)
sns.countplot(x = 'venue', data = matches, palette = 'magma')
plt.show()
```

```python
# The names of the top players in the Indian Premier League
top_players = matches.player_of_match.value_counts()[:10]
plt.figure(figsize = (10, 6))
fig, ax = plt.subplots()
ax.set_ylim([0, 20])
ax.set_ylabel('Count')
ax.set_title('Top player of the match winners')
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient = 'v', palette = 'flare')
plt.show()
```

```python
# The name of the batsman who scored the highest runs in the IPL
matches_run = deliveries['batsman'].value_counts().reset_index()
plt.figure(figsize = (10, 6))
matches_run.columns = ['batsman', 'batsman_runs']
plt.xticks(rotation = 90)
sns.barplot(x = 'batsman', y = 'batsman_runs',  data = matches_run[:25], palette = 'muted')
plt.show()
```

```python
# The counts of fours hit by the players
d4 = deliveries[deliveries['batsman_runs'] == 4]
batsman_4 = d4.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four', ascending = 0)
ax = batsman_4.iloc[:25,:].plot('batsman', 'four', kind = 'bar')
plt.title("The number of fours hit by players")
plt.xticks(rotation = 90)
plt.xlabel("Player name")
plt.ylabel("Number of fours")
plt.show()
```

```python
# The counts of sixes hit by the players
d6 = deliveries[deliveries['batsman_runs'] == 6]
batsman_6 = d6.groupby('batsman')['batsman_runs'].agg([('six', 'count')]).reset_index().sort_values('six', ascending = 0)
ax = batsman_6.iloc[:25, :].plot('batsman','six', kind = 'bar')
plt.title('The number of sixes hit by players')
plt.xticks(rotation = 90)
plt.xlabel('Player name')
plt.ylabel('Number of sixes')
plt.show()
```

```python
bat_first = matches[matches['win_by_runs'] != 0]
bat_first.head()
```

```python
sns.barplot(x = bat_first['winner'].value_counts().index, y = bat_first['winner'].value_counts().values)
plt.xticks(rotation = 90)
plt.title('The winner after batting first')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = deliveries['bowler'].value_counts().head(10).index, y = deliveries['bowler'].value_counts().head(10).values)
plt.title("The most played bowlers")
plt.show()
```

```python
# Players with most sixes
p = deliveries['batsman'][deliveries['total_runs'] >= 6].value_counts().head()
plt.figure(figsize = (10, 5))
sns.barplot(x = p.index, y = p.values)
plt.title('Players with the most sixes')
plt.show()
```

```python
deliveries['dismissal_kind'].value_counts()
```

```python
# The most common reason
plt.figure(figsize = (5, 15))
plt.title('The most common reason')
deliveries.groupby('dismissal_kind')['dismissal_kind'].count().plot(kind = 'pie')
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

```python
r = deliveries['bowler'][deliveries['dismissal_kind'] == 'bowled'].value_counts().head()
plt.figure(figsize = (10, 6))
plt.title('Bowlers who took the most wickets')
sns.barplot(x = r.index, y = r.values)
plt.show()
```

```python
s = deliveries['fielder'][deliveries['dismissal_kind'] == 'caught'].value_counts().head()
plt.figure(figsize = (10, 6))
sns.barplot(x = s.index, y = s.values)
plt.title("Fielders with the most catches")
plt.show()
```

```python
s = deliveries['fielder'][deliveries['dismissal_kind'] == 'run out'].value_counts().head()
plt.figure(figsize = (10, 6))
sns.barplot(x = s.index, y = s.values)
plt.title("Fielders who took the most wickets")
plt.show()
```

```python
s = deliveries['fielder'][deliveries['dismissal_kind'] == 'stumped'].value_counts().head()
plt.figure(figsize = (10, 5))
sns.barplot(x = s.index, y = s.values)
plt.title('Fielders who stumped the most')
plt.show()
```
